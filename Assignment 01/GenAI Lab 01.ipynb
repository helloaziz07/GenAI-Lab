Build a simple text generator using Markov Chains. 

```python
import random
from collections import defaultdict

class MarkovTextGenerator:
    def __init__(self):
        self.model = defaultdict(list)

    def train(self, text):
        words = text.lower().split()
        for i in range(len(words) - 1):
            self.model[words[i]].append(words[i + 1])

    def generate(self, start_word, length=30):
        start_word = start_word.lower()

        if start_word not in self.model:
            raise ValueError("Start word not found in training data")

        current_word = start_word
        generated = [current_word]

        for _ in range(length - 1):
            next_words = self.model[current_word]

            if not next_words:
                current_word = random.choice(list(self.model.keys()))
            else:
                current_word = random.choice(next_words)

            generated.append(current_word)

        return " ".join(generated)
```

```python
text = """
deep learning models learn patterns from large datasets
machine learning models improve with more data
data science combines statistics and programming
artificial intelligence creates smart applications
models trained on data can make predictions
learning algorithms optimize performance over time
"""
```

```python
# Train the generator

generator = MarkovTextGenerator()
generator.train(text)

print("Model trained successfully!")
```

```python
# Generate text

output1 = generator.generate(start_word="learning", length=12)
output2 = generator.generate(start_word="data", length=12)
output3 = generator.generate(start_word="models", length=12)

print(output1)
print(output2)
print(output3)
```
